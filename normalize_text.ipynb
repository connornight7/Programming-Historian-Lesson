```python
# html-to-list1.py
# We want to normalize some text by getting rid of any weird things it may have such as it capital letters or any punctial stuff attached
#We can use some tools from beautifulsoup to parse through the test and then get the text in its lower case form
import urllib.request
from bs4 import BeautifulSoup
import re

# 1. Load webpage
url = 'http://www.oldbaileyonline.org/browse.jsp?id=t17800628-33&div=t17800628-33'
response = urllib.request.urlopen(url)
html = response.read().decode('UTF-8')

# 2. Strip HTML tags
soup = BeautifulSoup(html, 'html.parser')
text = soup.get_text().lower()

# 3. Strip non-alphanumeric characters and split into words
def strip_non_alphanum(text):
    return re.findall(r'\b[a-z]+\b', text)  # keep only full lowercase words

wordlist = strip_non_alphanum(text)

print(wordlist)
```
